```python
def find_joint(img):
    h, w = img.shape
    im = img[int(h*0.5):int(h*0.75), :]
    cols = im.sum(axis = 1)
    cols = np.array(cols,dtype='int64')
    slope = []
    for i in range(0, len(cols)-1):
        slope.append(float(cols[i+1] - cols[i]))
    slope2 = []
    for i in range(0, len(slope)-11):
        slope2.append(sum(slope[i:i+11]))
    j = np.argmin(slope2) + int(h*0.5)
    return j
```

```python
def joint(img, j, m1 = 0.06, m2 = 0.06):
    h, w = img.shape
    im = img[int(j - h*m1):int(j + h*m2), :]
    return im
```

```python
def joint_only(im, threshold = 0.5):
    rows = im.sum(axis = 0)
    rows = np.array(rows,dtype='int64')
    norm = []
    for r in rows[:2800]:
        norm.append((r - min(rows))/ (max(rows) - min(rows)))
    nums = [i for i,v in enumerate(norm) if v > threshold]
    for i in range(len(nums)-1):
        diff = nums[i+1] - nums[i]
        if diff > 1:
            p = i   
    h, w = im.shape
    
    m1 = (h - (nums[p] - nums[0]))//2
    m2 = (h - (nums[p] - nums[0])) - m1
    right = im[:, nums[0] - m1 :nums[p] + m2]
    
    m1 = (h - (nums[-1] - nums[p+1]))//2
    m2 = (h - (nums[-1] - nums[p+1])) - m1
    left = im[:, nums[p+1] - m1 :nums[-1] + m2]
    return right, left 
```
